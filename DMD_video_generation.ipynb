### Setup

```python
from utils import DitherPainter, Frame
from itertools import product

save_dir = 'resources/'

dmd = Frame()
painter = DitherPainter(nrows=dmd.real_nrows, ncols=dmd.real_ncols, dither_method='Floyd-Steinberg')
```

### Generate various videos

#### A series of patterns of filled center cicle with decreasing radius

```python
start = 400
final = 20
step = 1

path = save_dir + f'video/dark_center_circle_with_decreasing_radius/start={start}_final={final}_step={step}/'

radius = range(start, final, -step)
for i, r in enumerate(radius):
    filename = f'{i}.bmp'

    dmd.drawPattern(painter.drawCircle(radius=r), color=0)
    dmd.saveDmdArrayToImage(path, filename, save_template=False)
```
