### Sequence based on circles
#### Black filled circles with decreasing radii

```python
r_start = 250
r_end = 20
r_step = 5

r_values = list(range(r_start, r_end - r_step, -r_step))
len(r_values)
```

```python
sequence_config = {
    'directory': save_dir + f'sequence/black_shrinking_circle_start={r_start}_end={r_end}_step={r_step}/',
    'filename': 'black_shrinking_circle',
    'description': 'Black circle on a white background with decreasing radius',
    'num_frames': 48,
    'params': [{
        'painter_method': 'drawCircle',
        'painter_args': {},
        'draw_args': {'color': 0},
        'sequence_varname': 'radius',
        'sequence_varvalues': r_values,
    }]
}

sequence = SequencePattern(**sequence_config)
sequence.save()
```

```python
sequence_config = {
    'directory': save_dir + f'sequence/black_moving_circle_start={x_start}_end={x_end}_step={x_step}/',
    'filename': 'black_moving_circle',
    'description': 'Black circle on a white background with moving center in x direction',
    'num_frames': 48,
    'params': [{
        'painter_method': 'drawCircle',
        'painter_args': {'radius': 20},
        'draw_args': {'color': 0},
        'sequence_varname': 'row_offset',
        'sequence_varvalues': x_values,
    }]
}

sequence = SequencePattern(**sequence_config)
sequence.save()
```
