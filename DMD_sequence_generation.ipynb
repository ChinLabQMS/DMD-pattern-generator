## Generate sequence of binary patterns and pack to RGB frames
### Setup

```python
from utils import BinarySequence, Painter

save_dir = 'resources/'
painter = Painter()
```

### A series of patterns with numbers, last pattern is black

From 1 to 96, 4 24-bit patterns

```python
sequence = BinarySequence(96)
for i in range(96): sequence.drawPatternOnFrame(i, painter.drawText(str(i+1)), color=1)
sequence.displayBinaryFrames()
sequence.saveSequenceToGIF(save_dir + '/packed_video/numbers/', 'numbers')
```

```python
sequence.packFramesToColor()
sequence.displayRGBFrames()
sequence.saveRGBFrames(save_dir + '/packed_video/numbers/', 'numbers', save_last_black=True)
```

### A series of white bands with increasing widths, first pattern is black

```python
num = 96
sequence = BinarySequence(num)

for i in range(1, num):
    sequence.drawPatternOnFrame(i, painter.drawAngledLine(row_offset=0, width=10*i, center=True), color=1)
sequence.saveRGBFrames(save_dir + '/packed_video/expanding_whiteband/', 'whiteband')
sequence.saveBinaryFrames(save_dir + '/packed_video/expanding_whiteband/binary/', 'whiteband')
sequence.saveSequenceToGIF(save_dir + '/packed_video/expanding_whiteband/', 'whiteband')
```

### A series of filled dark circles with decreasing radii

```python
r_start = 260
r_end = 20
r_step = 5

r_range = list(range(r_start, r_end, -r_step))
num_frames = len(r_range)
num_frames
```

```python
sequence = BinarySequence(num_frames)

for i, r in enumerate(r_range):
    sequence.drawPatternOnFrame(i, painter.drawCircle(radius=r), color=0)
sequence.saveRGBFrames(save_dir + f'/packed_video/shrinking_dark_circle_start-{r_start}_end-{r_end}_step-{r_step}/', 'shrinking_dark_circle')
sequence.saveBinaryFrames(save_dir + f'/packed_video/shrinking_dark_circle_start-{r_start}_end-{r_end}_step-{r_step}/binary/', 'shrinking_dark_circle')
sequence.saveSequenceToGIF(save_dir + f'/packed_video/shrinking_dark_circle_start-{r_start}_end-{r_end}_step-{r_step}/', 'shrinking_dark_circle')
```
