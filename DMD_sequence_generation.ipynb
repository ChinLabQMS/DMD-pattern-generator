## Generate sequence of binary patterns and pack to RGB frames
### Setup

```python
from utils import BinarySequence, Painter

save_dir = 'resources/'
painter = Painter()
```

### A series of patterns with numbers

From 1 to 96, 4 24-bit patterns

```python
sequence = BinarySequence(96)
for i in range(96): sequence.drawPatternOnFrame(i, painter.drawText(str(i+1)), color=1)
sequence.displayBinaryFrames()
sequence.saveSequenceToGIF(save_dir + '/packed_video/numbers/', 'numbers')
```

```python
sequence.packFramesToColor()
sequence.displayRGBFrames()
sequence.saveRGBFrames(save_dir + '/packed_video/numbers/', 'numbers', save_last_black=True)
```

### A series of white bands with increasing widths

```python
sequence = BinarySequence(96)
for i in range(96): sequence.drawPatternOnFrame(i, painter.drawAngledLine(row_offset=0, width=10*(i+1), center=True), color=1)
sequence.displayBinaryFrames()
sequence.saveSequenceToGIF(save_dir + '/packed_video/whiteband/', 'whiteband')
```

```python
sequence.packFramesToColor()
sequence.displayRGBFrames()
sequence.saveRGBFrames(save_dir + '/packed_video/whiteband/', 'whiteband', save_last_black=True)
```

### A series of filled dark circles with decreasing radii

```python
r_start = 260
r_end = 20
r_step = 5

folder_name = save_dir + f'/packed_video/shrinking_dark_circle_start-{r_start}_end-{r_end}_step-{r_step}/'
filename = 'shrinking_dark_circle'

r_range = list(range(r_start, r_end, -r_step))
num_frames = len(r_range)
num_frames
```

```python
sequence = BinarySequence(num_frames)
for i, r in enumerate(r_range): sequence.drawPatternOnFrame(i, painter.drawCircle(radius=r), color=0)
sequence.saveRGBFrames(folder_name, filename)
sequence.saveBinaryFrames(folder_name + '/Binary/', filename)
sequence.saveSequenceToGIF(folder_name, filename)
```

### A series filled 2x2 dark circles with decreasing radii

```python
r_start = 70
r_end = 22
r_step = 2
anchor = ((-50, 50), (-50, -50), (50, 50), (50, -50))

folder_name = save_dir + f'/packed_video/shrinking_dark_anchor_start-{r_start}_end-{r_end}_step-{r_step}/anchor={anchor}/'
filename = 'shrinking_dark_anchor'

r_range = list(range(r_start, r_end, -r_step))
num_frames = len(r_range)
num_frames
```

```python
sequence = BinarySequence(num_frames)
for i, r in enumerate(r_range): sequence.drawPatternOnFrame(i, painter.drawAnchorCircles(anchor=anchor, radius=r), color=0)
sequence.saveRGBFrames(folder_name, filename)
sequence.saveBinaryFrames(folder_name + '/Binary/', filename)
sequence.saveSequenceToGIF(folder_name, filename)
```
